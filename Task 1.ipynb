```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_40826.csv", skiprows=4)
df.head()
```

Check Columns

```python
df.columns
```

Get India 2022 Population

```python
india = df[df["Country Name"] == "India"]

india_2022 = india["2022"].values[0]

print("India Population in 2022:", india_2022)
```

Create Bar Chart

```python
countries = ["India", "China", "United States"]

filtered = df[df["Country Name"].isin(countries)]

pop_2022 = filtered[["Country Name", "2022"]]

plt.figure()
plt.bar(pop_2022["Country Name"], pop_2022["2022"])

plt.xlabel("Country")
plt.ylabel("Population (2022)")
plt.title("Population Comparison (2022)")
plt.show()
```
